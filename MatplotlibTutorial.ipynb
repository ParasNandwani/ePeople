## Scatter plot to show clusters
while we have seen a  scatter  plot of random dots before, scatter plots are very useful with representing discrete data points that show a trend and clusters .Each data series will be plotted in different color per plottig command by default,which helps us distinguish the distinct dots from each series


```
np.random.seed(123)

x_0=np.random.rand(10)+1.5


y_0=np.random.rand(10)+2


x_1=np.random.rand(10)+2.5


y_1=np.random.rand(10)+2

plt.scatter(x_0,y_0)
plt.scatter(x_1,y_1)
plt.show()
```

We can observe from above plant there are two artificially created clusters of dots (left half) and orange(right half).   
There is another way to generate clusters and plot them on scatter plots.For this ,we make use of make_blobs() function found in sklearn package

# Adusting axes,grids,labels,titles and legends'
Good visualisations often requires design input to suit our custom data visualisation needs and purposes.Text labels are needed to make figures informative in many cases.

```
d=[11,12,13,14,15,16,17]
t0=[15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1=[26.1,26.2,24.3,25.1,12.3,26.7,27.8]

plt.plot(d,t0)
plt.plot(d,t1)
plt.xlim(11,15)
plt.ylim(0,30)
plt.show()
```

**Adding axis labels**
To give meaning to the values on the x and y axis with add axis lables with plt.xlabel() or plt.ylabel()

```
d=[11,12,13,14,15,16,17]
t0=[15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1=[26.1,26.2,24.3,25.1,12.3,26.7,27.8]

plt.plot(d,t0)
plt.plot(d,t1)
plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12,fontweight='semibold')
plt.ylabel('Temperature (c)',size=12,fontweight='semibold')
plt.show()
```

**Adding a grid**.  
Whiile a blank plot background is clean,sometimes we may like to get some reference gridlines for better reference in multiline cases.We can turn on the background gridlines by calling plt.grid(True) before plt.show()

```
d=[11,12,13,14,15,16,17]
t0=[15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1=[26.1,26.2,24.3,25.1,12.3,26.7,27.8]

plt.plot(d,t0)
plt.plot(d,t1)
plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12,fontweight='semibold')
plt.ylabel('Temperature (c)',size=12,fontweight='semibold')
plt.grid(True,linewidth=0.5)
plt.show()
```

**Adding a Title plt.title('Title')**.   
Inorder to describe the information of the plotted data, we can dive a title to the figure.This can be done using plt.title('title)

```
d=[11,12,13,14,15,16,17]
t0=[15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1=[26.1,26.2,24.3,25.1,12.3,26.7,27.8]

plt.plot(d,t0)
plt.plot(d,t1)
plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12,fontweight='semibold')
plt.ylabel('Temperature (c)',size=12,fontweight='semibold')
plt.grid(True,linewidth=0.5)
plt.title("Daily temperature of 2 cities",size=14,fontweight="bold")
plt.show()
```

**Addiing a legend**.  
To match a data series on plots with their labels,such as by their line styles and marker styles  we add plt.legend()

```
d=[11,12,13,14,15,16,17]
t0=[15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1=[26.1,26.2,24.3,25.1,12.3,26.7,27.8]

plt.plot(d,t0)
plt.plot(d,t1)
plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12,fontweight='semibold')
plt.ylabel('Temperature (c)',size=12,fontweight='semibold')
plt.grid(True,linewidth=0.5)
plt.title("Daily temperature of 2 cities",size=14,fontweight="bold")
plt.legend(['Tokyo','HAWAAI'])
plt.show()
```

# Native Matplotlib Plot Types    
Besides the most basic scatter plot and line plots.Matplotlibs provides a versatile collection of plot types to serve different data visualisation purposes.

In choosing the right plot type to display we have the following considerations:
- Number of variables 
- Distribution of data
- Relationship between data series

**Histogram**.   
Histogram are useful in surveying the distribution of data.For example we can plot data on a histogram when we want to see some age groups distributed in a population,light exposure in a photograph or the amount of precipitation in each month of a city.    
In matplotlib,we call the plt.hist() with a linear array.Matplotlib will automatically group them to bins and plot out the frequencies for each bin in bars

```
x=np.random.binomial(100,0.5,size=10000)
print(len(x))
plt.hist(x)
plt.show()
```

**Bar Plot**
Bar plots are useful for comparing absolute levels of discrete data series.

```
# let's take an example of market capitalization of today's   much hyped cryptocurrencies
# Data series consists of top 5 cryptocurrencies in terms of maket capitalization

cc=['BTC','XRP','ETH','BCH','ADA']
cap=[282034,131378,10793,49999,26137]

plt.bar(cc,cap)
plt.title('Market Capitalization Of Five Top of cryptocurrencies')
plt.xlabel('Cryptocurrency')
plt.ylabel('Market Capitalization (million usd)')
plt.show()
```

To create bar plots with bars in the designated order,we can make use of pandas and matplotlib integration.The procedure is as follow as :
1. Create a Pandas Dataframe df
2. Plot the bar chart with df.plot(kind='bar')
3. Set the labels of xticks
4. Adjust other plot properties
5. Show the plot with plt.show()

```
import pandas as pd

df=pd.DataFrame({'cc':cc,'cap':cap})
ax=df.plot(kind='bar')
ax.set_xticklabels(df['cc'])
plt.show()
```

**Pie Chart**.  
Pie chart is a circular representation of component ratios.The angle and hence the arc length each sector ration(called 'wedges') presents tne porition that each component accounts for ,relative to the whole.Matplotlb uses plt.pie() function to draw pie charts.We can label each sector with labels 

```
labels=['nginx','apache','IIS','varnish','others']
x=[0.31,0.3,0.14,0.1,0.15]
plt.pie(x,labels=labels,autopct='%1.1f%%')
plt.title('Web server usage stats')
plt.show()
```
